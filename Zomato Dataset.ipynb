```python
User_rating=Final_dataset.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Counting'})
```

```python
User_rating
```

## Observation
After using Zomato, most of the customers are not even rated the service.

```python
sns.distplot(User_rating['Aggregate rating'])
```

```python
plt.figure(figsize=[12,6])
sns.barplot(x='Aggregate rating',y='Counting',data=User_rating,hue='Rating text')
```

## Observation
Majority of the customers give the average and good rating.

```python
Final_dataset.groupby(['Aggregate rating','Country']).size().reset_index()
```

## Observation

```python
Highest_country_rating=Final_dataset[Final_dataset['Aggregate rating']>=4.5].groupby([Final_dataset['Country']]).size().reset_index().rename(columns={0:'Count'})
Highest_country_rating
```

```python
plt.figure(figsize=[18,7])
sns.barplot(x='Country',y='Count',data=Highest_country_rating)
```
